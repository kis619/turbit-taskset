```python
all_entries = collection1.find()
desired_columns = ["Dat/Zeit", "Wind (m/s)", "Leistung (kW)"]
date_wind_dict = {}
while all_entries.alive:
    entry = all_entries.next()
    date_wind_dict[entry.get(desired_columns[0])] = [
        entry.get(desired_columns[1]),
        entry.get(desired_columns[2])
    ]
```

```python
print(date_wind_dict)
```

```python
date_wind_df = pd.DataFrame(date_wind_dict, index=["wind", "power"])
```
